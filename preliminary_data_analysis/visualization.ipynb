```python
import pandas as pd 
import numpy as np

df1 = pd.read_csv('./scaled_PCA_weighted.csv')
df2 = pd.read_csv('./ward_WI_unweightedPCA.csv')
```

```python
df1['year']
```

```python
df2['year_code'] = np.where(df2['year'] == 2019, '2019_2', '2014')
df1['year_code'] = np.where(df1['year'] == 2019, '2019_1', '2016')
df = pd.concat([df1, df2])
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Set the style of seaborn for better visibility
sns.set(style="whitegrid")

# Create a density plot for each year
plt.figure(figsize=(10, 6))
sns.kdeplot(data=df, x="mean_pos_WI", hue="year_code", fill=True, common_norm=False, alpha=0.5)

plt.title('Density Distribution of mean_pos_WI by Year')
plt.xlabel('mean_pos_WI')
plt.ylabel('Density')
plt.show()
```
